```python
import pandas as pd

url_2020 = 'https://www.pro-football-reference.com/years/2020/fantasy.htm#fantasy'
url_2019 = 'https://www.pro-football-reference.com/years/2019/fantasy.htm#fantasy'
url_2018 = 'https://www.pro-football-reference.com/years/2018/fantasy.htm#fantasy'
url_2017 = 'https://www.pro-football-reference.com/years/2017/fantasy.htm#fantasy'

#This function cleans the data
def clean_data(url):
    dfs = pd.read_html(url, header=1)
    df_holder = dfs[0]
    df_holder = df_holder[df_holder.Rk != "Rk"]
    df_holder = df_holder[["Player","Age","G","Tm","FantPos","FantPt","PosRank"]]
    df_holder[['Age','G','FantPt','PosRank']] = df_holder[['Age','G','FantPt','PosRank']].apply(pd.to_numeric)                                     
    df_holder["Player"] = df_holder["Player"].str.replace(' +*','', regex = False)
    df_holder["Player"] = df_holder["Player"].str.replace(' *','', regex = False)
    df_holder["Player"] = df_holder["Player"].str.replace(' +','', regex = False)
    df_holder["Player"] = df_holder["Player"].str.replace('*','', regex = False)
    df_holder["Player"] = df_holder["Player"].str.replace('+','', regex = False)
    
    df_rb = df_holder[df_holder.FantPos == "RB"]
    df_rb = df_rb.head(24)
    
    df_wr = df_holder[df_holder.FantPos == "WR"]
    df_wr = df_wr.head(24)
    
    df_te = df_holder[df_holder.FantPos == "TE"]
    df_te = df_te.head(12)
    
    df_qb = df_holder[df_holder.FantPos == "QB"]
    df_qb = df_qb.head(12)
    
    list_holder = [df_holder,df_rb,df_wr,df_te,df_qb]
    return list_holder

#This function seperates into rb
#def rb_data(df_cleaned):
    #df_rb = df_cleaned[df_cleaned.FantPos == "RB"]
    #df_rb = df_rb.head(36)
    #return df_rb

    
    
```
